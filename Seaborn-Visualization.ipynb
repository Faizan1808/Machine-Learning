```python
# import packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# importing seaborn
import seaborn as sns
```

```python
sns.__version__
```

```python
df = sns.load_dataset('tips') # loading the tips dataset from seaborn package
df.head()
```

```python
gender_num = df.sex.value_counts()
gender_num
```

```python
print(type(gender_num))
```

```python
s = pd.Series([1,2,3,4,5],index=['A','B','C','D','E'])
s.index
```

```python
plt.bar(x=gender_num.index,height=gender_num)
plt.title('Count of Males and females')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
sns.countplot(df.sex)
plt.title('Count of Males & Females')
plt.show()
```

```python
avg_bill = df.groupby('sex')['total_bill'].mean()
```

```python
sns.barplot(x = avg_bill.index,y = avg_bill)
```

```python
plt.hist(df.total_bill,bins=10,edgecolor='k')
plt.title('Distribution of total bill')
plt.xlabel('Total bill')
plt.ylabel('Frequency')
plt.show()
```

```python
# sns.histplot(df.total_bill,hue=)
# plt.show()
```

```python
sns.distplot(df.total_bill)
plt.show()
```

```python
sns.countplot(x='day',data=df,hue='sex')
plt.show()
```

```python
# plot boxplot
sns.boxplot(y=df.total_bill)
plt.show()
```

```python
sns.boxplot(x='smoker',y='total_bill',data=df,hue='sex')
plt.show()
```

```python
sns.pairplot(df)
plt.show()
```

```python
sns.violinplot(x='smoker',y='total_bill',data=df,hue='sex')
plt.show()
```
